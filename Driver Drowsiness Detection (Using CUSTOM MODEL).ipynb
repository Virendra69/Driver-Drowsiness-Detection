```python
import cv2
import tensorflow as tf
import numpy as np
```

```python
model = tf.keras.models.load_model("E:\\VS Code Programs\\Python_Codes\\OpenCV\\model")
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    _, frame = cap.read()
    
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    frame_gray = cv2.resize(frame_gray, (50, 50))
    frame_gray_array = np.array(frame_gray)
    frame_gray_array = np.expand_dims(frame_gray_array, axis = 0)
    
    res = model.predict([frame_gray_array])
    
    if res == 1:
        cv2.putText(frame, 'Drowsy', (100, 100), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    else:
        cv2.putText(frame, 'Not Drowsy', (100, 100), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    
    cv2.imshow('FRAME', frame)
    if cv2.waitKey(1) == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
